```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
import string
from sklearn.feature_extraction.text import CountVectorizer

%matplotlib inline
```

```python
# read data, and concat

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
train = pd.concat([train,test])



```

```python
# transform reg_date to number, and then normalize
train['reg_date'].fillna(train.original_reg_date.str[-4:],inplace=True)

# string to int
train['reg_date'] = train.reg_date.str[-4:].astype(int)

train['reg_date'] = 2021-train['reg_date']

```

```python
# these features are filled by average number

# power
mean_power = train.power.mean()
train.power.fillna(mean_power,inplace=True)

# engine_cap
mean_engine_cap = train.engine_cap.mean()
train.engine_cap.fillna(mean_engine_cap,inplace=True)

# mileage
median_mileage = train.mileage.median()
train.mileage.fillna(median_mileage,inplace=True)
# print(train.mileage[5])


mean_curb_weight = train.curb_weight.median()
train.curb_weight.fillna(mean_curb_weight,inplace=True)

# no_of_owners 
train.no_of_owners.fillna(1,inplace=True)

```

```python
# function to find the mean value according to 'make' and 'model'

def get_mean_value(train,make,model,feature):
   
    mean_val = train.loc[(train.make == make) & (train.model == model)][feature].dropna().mean()
    
    if np.isnan(mean_val):
        mean_val = train.loc[(train.make == make)][feature].dropna().mean()
        if np.isnan(mean_val):
            mean_val = train[feature].dropna().mean()         
    return mean_val

# get_mean_value(train,'mercedes-benz','300sl','depreciation')
# get_mean_value(train,'bmw','x6','coe')
```

```python

# function to feature NaN value according to features
def fill_nan(train,feature):
    for index, row in train.iterrows():
        if np.isnan(row[feature]): 
            train[feature][index] = get_mean_value(train,row.make,row.model,feature)


```

```python

# depreciation
fill_nan(train,'depreciation')
#     print(train.depreciation.isna().sum())
#     print(train.depreciation[7])

# coe
fill_nan(train,'coe')
# print(train.coe.isna().sum())
# print(train.coe[36])

# dereg_value
fill_nan(train,'dereg_value')
# print(train.dereg_value.isna().sum())

# omv
fill_nan(train,'omv')
# print(train.omv.isna().sum())

fill_nan(train,'road_tax')
```

```python
# calculate arf by omv
def cal_arf(train):
    for index, row in train.iterrows():
#     print(row)
        if np.isnan(row['arf']):
            if row['omv']<=20000:
                train['arf'][index] = row['omv']
            elif row['omv']<=50000:
                train['arf'][index] = 20000 + (row['omv']-20000)*1.4
            elif row['omv'] >50000:
                train['arf'][index] = 20000 +30000*1.4 + (row['omv']-50000)*1.8
```

```python
# arf
cal_arf(train)

```
